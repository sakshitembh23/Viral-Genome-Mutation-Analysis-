```python

record = SeqIO.read("MS2_phage_seq.fasta", "fasta")
print(record.seq)       #record.seq will only print the seq part of the file.
```

```python
# defining variables
input_seq = record.seq              #using the seq of the file as a input seq for further processing.
start_codon = "ATG"
stop_codons = ["TAG","TGA","TAA"]
ORF_num = 0
frame_num = 1
ORF_begin_loc = 0
ORF_end_loc = 0
ORF_len = 75                                                                     #subject to change 

#making empty dictionary for storing ORF sequences. 
ORF_seqs = {}

#dividing the sequence into 3 by creating subsets of a string.
nt_pos=0          #initial value is 0.
while nt_pos <(len(input_seq)-2):
##for nt_pos in range(0,(len(input_seq)-2),3):                                  #range(start,stop,step)
                                              
  read_codon=input_seq[nt_pos: nt_pos+3]                                        #storing the values into another temporary variablea
  if read_codon==start_codon:                                                   #stating the condition that the read codon is equal to the start codon or not, if it is proceed further.
    for new_pos in range(nt_pos+3,(len(input_seq)-2),3):                        #storing the new values
      read_codon2=input_seq[new_pos:new_pos+3]
      if read_codon2 in stop_codons:
        if (new_pos-nt_pos) >= ORF_len:       #adding length cut off for orf sequence. Sequences of length greater than or equal to the ORF_len will only be selected
          ORF_begin_loc=nt_pos                #storing orf begin and end locations in the new introduced variables
          ORF_end_loc=new_pos
          ORF_num= ORF_num+1                  #increasing the orf number by one in each cycle. 
          #ORF_seq=input_seq[ORF_begin_loc:(ORF_end_loc+3)] #here the orf_seq which was dict has now chnaged to a string as we are storing all the for loop values in it.
          ORF_seqs["ORF"+str(ORF_num)]=input_seq[ORF_begin_loc:(ORF_end_loc+3)] #storing values directly to the dict ##["ORF"+str(ORF_num)]is the key and input_seq[ORF_begin_loc:(ORF_end_loc+3)] is the value.
          nt_pos=new_pos+3                    # to prevent loop to again start with the same location we are saying that for every cycle add three to the start position location.  
        break                                  # if we find a stop codon then we need to get out of the loop thus use break to break the loop and not continue the process.    

  nt_pos=nt_pos+3                              # even if we don't find the stop codon we still want it to restart again.

# printing the orf begin and end location, orf number and orf sequences.                                           
print(ORF_begin_loc)
print(ORF_end_loc)
print(ORF_num)
#print(ORF_seq)
print(ORF_seqs)

    

```
